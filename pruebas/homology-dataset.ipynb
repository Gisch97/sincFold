```python
l = (dataset.id.str.split('_').apply(lambda x: x[0]))
dataset['family'] = l
dataset
```

```python
dataset[dataset['family']!= 'tRNA'].drop(columns='family')
```

```python
family = ['tRNA', 'RNaseP', 'telomerase']
for fam in family:
    train   = dataset[dataset['family'] != fam].drop(columns='family')
    train.to_csv(f'train_{fam}.csv')
    test    = dataset[dataset['family'] == fam].drop(columns='family')
    train.to_csv(f'test_{fam}.csv')
```
